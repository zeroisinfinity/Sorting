```python
import cProfile;import random;import math

def shuffle(array,start_pos,end_pos):
    sh_pivot=random.randint(start_pos,end_pos)
    array[sh_pivot],array[start_pos]=array[start_pos],array[sh_pivot]
    return partition(array,start_pos,end_pos)
def partition(array,start_pos,end_pos):
    pivot=array[end_pos]
    minimum=start_pos-1
    for index in range(start_pos,end_pos):
        if array[index]<pivot:
            minimum+=1
            array[minimum],array[index]=array[index],array[minimum]
    array[minimum+1],array[end_pos]=array[end_pos],array[minimum+1]
    return minimum+1
def quick_sort(array,start_pos,end_pos):
    if start_pos<end_pos:
        current_pivot=shuffle(array,start_pos,end_pos)
        
        quick_sort(array,start_pos,current_pivot-1)
        quick_sort(array,current_pivot+1,end_pos)
arr=[9,7,6,5,4,2,7,8,9,2,5]*70+[900]*80+list(range(90000,578,-6))
cProfile.run("quick_sort(arr,0,len(arr)-1)")

#print(arr)
```

```python
import sys
import random
import math
sys.setrecursionlimit(10000000)
def quick_sort(array, start_pos, end_pos):
    new_pivot = 0
    if start_pos < end_pos:
        new_pivot = shuffle(array, start_pos, end_pos)
        quick_sort(array, start_pos, new_pivot)
        quick_sort(array, new_pivot + 1, end_pos)
def shuffle(array,start_pos,end_pos):
    sh_pivot = random.randint(start_pos,end_pos)
    array[sh_pivot] , array[start_pos] = array[start_pos] , array[sh_pivot]
    return partition(array,start_pos,end_pos)
def partition(array, start_pos, end_pos):
    pivot = array[start_pos]
    inc = start_pos - 1
    dec = end_pos + 1
    while True:
        inc += 1
        while array[inc] < pivot:
            inc += 1
        dec-=1
        while array[dec] > pivot:
            dec-=1
        if inc >= dec:
            return dec
        array[inc] , array[dec] = array[dec] , array[inc]
arr = [9, 7, 6, 5, 4, 2, 7, 8, 9, 2, 5]*4999 + [900] * 800 + list(range(90000, 578, -6))
quick_sort(arr,0,len(arr) - 1)
cProfile.run("quick_sort(arr,0,len(arr)-1)")
```

```python
def iterative_bubble_sort(array):
    for limiter in range(len(array)):
        swap_flag = False
        for cf in range(0, len(array) - limiter - 1):
            if array[cf] > array[cf + 1]:
                array[cf], array[cf + 1] = array[cf + 1], array[cf]
                swap_flag = True
        if swap_flag == False:
            break
    return array



arr = [9,7,6,5,4,2,7,8,9,2,5]*4999 + [900]*800 + list(range(90000,578,-6))
cProfile.run("iterative_bubble_sort(arr)")
```

```python
import sys
sys.setrecursionlimit(100000)

def insertion_sort(array):

    for test_sub in range(1,len(array),1):
            sample = array[test_sub]
            compare = test_sub - 1
            while compare >= 0 and sample < array[compare]:
                array[compare + 1] = array[compare]
                compare -= 1
            array[compare+1] = sample
    return array

def merge_sort(array,start_pos,mid,end_pos):
    merged_arr = [0] * ( end_pos - start_pos + 1 )
    index_mrg = 0
    index_i = start_pos
    index_j = mid+1
    if len(merged_arr) < 10:
        merged_arr = insertion_sort(merged_arr)
    while index_i <= mid  and  index_j <= end_pos :
        if array[index_i] < array[index_j]:
            merged_arr[index_mrg] = array[index_i]
            index_i+=1
        else:
            merged_arr[index_mrg] = array[index_j]
            index_j+=1
        index_mrg+=1
    while index_i <= mid :
        merged_arr[index_mrg] = array[index_i]
        index_mrg+=1
        index_i+=1
    while index_j <= end_pos:
        merged_arr[index_mrg] = array[index_j]
        index_mrg += 1
        index_j += 1
    for update in range(len(merged_arr)):
        array[start_pos + update] = merged_arr[update]

def binary_cut(array,start_pos,end_pos):
    if start_pos<end_pos:
        mid = start_pos + (end_pos - start_pos)//2
        binary_cut(array , start_pos , mid)
        binary_cut(array , mid + 1 , end_pos)
        merge_sort(array,start_pos,mid,end_pos)

arr = [9,7,6,5,4,2,7,8,9,2,5]*4999 + [900]*800 + list(range(90000,578,-6))

cProfile.run("binary_cut(arr,0,len(arr)-1)")
```
